```python
all_tables=soup.find_all('table') # it will find all the tables in the HTML
all_tables
```

```python
right_table=soup.find('table',class_='wikitable sortable') # here we are using find() method to find the necessary data by taking class variable
right_table
```

```python
# creating empty lists to store the table contents
# In our table there are 3 columns.
# we are using 3 lists to store each column data in a seperate lists
A=[]
B=[]
C=[]
```

```python
# getting the table contents by using for loop
# tr contains all table rows
# td contains all table contents
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        
```

```python
# second column in the table contains some missing values 
# replace them with NaN.so that we can drop those using dropna() method
for b in range(len(B)):
    if B[b]=='Not assigned\n':
        B[b]='NaN'
```

Transform data into pandas dataframe

```python
# creating a data frame 
df=pd.DataFrame(A,columns=['PostalCode'])
```

```python
# adding 3 lists to data frame
df['borough']=B
df['neighborhood']=C
```

```python
# replacing nan values 
nan_value=float('NaN')
df.replace('NaN',nan_value,inplace=True)
```

```python
# dropping all nan values using dropna() method
df.dropna(subset=['borough'],inplace=True)
```

```python
# creating new index values 
index=list(range(1,104))
df.index=index
df.head(10)
```

```python
# replacing '/' and '\n' in neighborhood column
df['neighborhood']=df['neighborhood'].str.replace('/',',')
df['neighborhood']=df['neighborhood'].str.replace('\n','')
```

```python
# create a function that repeats the same process for all neighborhoods
import requests
def getNearbyVenues(names,latitude,longitude,radius=500):
    venues_list=[]
    for name,lat,lng in zip(names,latitude,longitude):
        print(name)
        url='https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID,CLIENT_SECRET,VERSION,lat,lng,radius,limit)
        results=requests.get(url).json()['response']['groups'][0]['items']
        venues_list.append([(name,lat,lng,v['venue']['name'],v['venue']['location']['lat'],v['venue']['location']['lng'],v['venue']['categories'][0]['name']) for v in results])
    nearby_venues=pd.DataFrame([venue for venue_list in venues_list for venue in venue_list])
    nearby_venues.columns=['Neighborhood','Neighborhood Latitude','Neighborhood Longitude','Venue','Venue Latitude','Venue Longitude','Venue Category']
    return(nearby_venues)
```

# Analyse Each Neighborhood

```python
# here we are using one hot coding to get dummy values for each venue
toronto_onehot=pd.get_dummies(Toronto_Venues[['Venue Category']],prefix='',prefix_sep='')
# getting back neighborhood column back to data frame
toronto_onehot['Neighborhood']=Toronto_Venues['Neighborhood']
# moving the neighborhood column to the first
fixed_columns=[toronto_onehot.columns[-1]]+list(toronto_onehot.columns[:-1])
toronto_onehot=toronto_onehot[fixed_columns]
toronto_onehot.tail(15)
```

Now group the rows by neighborhood and by taking mean and frequency for each category 

```python
toronto_grouped=toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
# lets find the size of the above table
toronto_grouped.shape
```

Now let's print all the nieghborhood with top-5 venues

```python
num_top=5
for hood in toronto_grouped['Neighborhood']:
    print('------'+hood+'-------')
    temp=toronto_grouped[toronto_grouped['Neighborhood']==hood].T.reset_index()
    temp.columns=['venue','freq']
    temp=temp.iloc[1:]
    temp['freq']=temp['freq'].astype(float)
    temp=temp.round({'freq':2})
    print(temp.sort_values('freq',ascending=False).reset_index(drop=True).head(num_top))
    print('\n')
```

Now put them into a new data frame

```python
# sort them in descending order
def most_common_venues(row,num_top):
    row_categories=row.iloc[1:]
    row_categories_sorted=row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top]
```

Now create a new data frame and put top-10 venues of each neighborhood

```python
num_top_venues=10
columns=['Neighborhood']
indicators=['st','nd','rd']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} most commom venues.'.format(ind+1,indicators[ind]))
    except:
        columns.append('{} most common venue'.format(ind+1))
neighborhood_venues_sorted=pd.DataFrame(columns=columns)
neighborhood_venues_sorted['Neighborhood']=toronto_grouped['Neighborhood']
for ind in np.arange(toronto_grouped.shape[0]):
    neighborhood_venues_sorted.iloc[ind,1:]=most_common_venues(toronto_grouped.iloc[ind,1:],num_top_venues)
neighborhood_venues_sorted.head()
```

## Cluster Neighborhood

Run the kmeans to cluster the neighborhood into 5 clusters

```python
# set no of clusters
kclusters=5
toronto_clustering=toronto_grouped.drop('Neighborhood',1)
# run kmean clustering
kmeans=KMeans(n_clusters=kclusters,random_state=0).fit(toronto_clustering)
# check out cluster labels generated for each row in the data frame
kmeans.labels_[0:10]
```

Now let's create a new data frame that includes all neighborhood with top-10 venues and their clusters

```python
# add clustering labels
neighborhood_venues_sorted.insert(0,'Cluster Labels',kmeans.labels_)
toronto_merged=toronto
#merge toronto groouped with toronto data to add latitude and latitude and longitude columns
toronto_merged=toronto_merged.join(neighborhood_venues_sorted.set_index('Neighborhood'),on='neighborhood')
toronto_merged.head()
```

Now finally visualise the resulting clusters in map

```python
map_clusters=folium.Map(location=[latitude,longitude],zoom_start=12)
x=np.arange(kclusters)
ys=[i+x+(i*x)**2 for i in range(kclusters)]
color_array=cm.rainbow(np.linspace(0,1,len(ys)))
rainbow=[colors.rgb2hex(i) for i in color_array]
markers_colors=[]
for lat,lng,poi,cluster in zip(toronto_merged['latitude'],toronto_merged['longitude'],toronto_merged['neighborhood'],toronto_merged['Cluster Labels']):
    label=folium.Popup(str(poi)+'Cluster'+str(cluster),parse_html=True)
    folium.CircleMarker([lat,lng],
                        radius=5,
                        popup=label,
                        color=rainbow[cluster-1],
                        fill=True,
                        fill_color=rainbow[cluster-1],
                        fill_opacity=0.7).add_to(map_clusters)
map_clusters
```
